```python
def merge_datasets(df1, df2, date_column):
    merged_df = pd.merge(df1, df2, on=date_column, how='inner')
    return merged_df
# "날짜"가 같으면 merge
```

```python
merge_df = merge_datasets(prec_df, temp_df, '날짜' )
merge_df
```

```python
def merge_datasets_by_date(df1, df2, datetime_column, date_column):
    df1[date_column] = pd.to_datetime(df1[date_column]).dt.date
    df2[datetime_column] = pd.to_datetime(df2[datetime_column]).dt.date
    merged_df = pd.merge(df1, df2, left_on=date_column, right_on=datetime_column, how='inner')
    return merged_df

# 날짜가 같으면 merge
```

```python
print(merge_df.columns)
```

```python
merged_data = merge_datasets_by_date(merge_df, wc_df, '경기(행사) 일시', '날짜')
merged_data
```
